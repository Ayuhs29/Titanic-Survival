```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data import and details

```python
df_train = pd.read_csv('titanic/train.csv',index_col='PassengerId')
df_test = pd.read_csv('titanic/test.csv',index_col='PassengerId')
```

## EDA

```python
sns.heatmap(df_train[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr(),cmap="magma")
plt.show()
```

```python
survived=df_train['Survived']
df_train.drop(['Survived'],axis=1,inplace=True)
df_train
```

```python
combined=pd.concat([df_test,df_train])
combined
```

```python
combined.isnull().sum()
```

```python
y= combined['Age'].value_counts()
x= y.index.values
plt.bar(x,y)
plt.title('Age based division')
plt.show()


for i in range(1,4):
    y=combined[combined['Pclass']== i]['Age'].value_counts()
    x=y.index.values
    plt.bar(x,y)
    plt.title('Class'+str(i))
    plt.show()


median=combined.groupby('Pclass')['Age'].median()
print(median)

```

Age missing val 

```python
combined['Age'] = combined['Age'].fillna(combined['Pclass'].map(median))

combined.isnull().sum()
```

Fare missing Val

```python
combined[combined['Fare'].isnull()]

fare_median=combined.groupby(['Pclass','Embarked'])['Fare'].median()

combined['GroupKey'] = list(zip(combined['Pclass'], combined['Embarked']))
combined['Fare'] = combined['Fare'].fillna(combined['GroupKey'].map(fare_median))
combined.drop(columns='GroupKey', inplace=True)

combined.isnull().sum()
```

Embarked missing val

```python
combined[combined['Embarked'].isnull()]
combined[(combined['Pclass']==1) & (combined['Fare']<513) & (combined['Fare']>=73)]['Embarked'].value_counts()

combined['Embarked']=combined['Embarked'].fillna('C')
combined.isnull().sum()
```

```python
combined.drop(columns='Cabin', inplace=True)
combined.isnull().sum()
```

## Feature selection


```python
combined.reset_index(inplace=True)
```

```python
combined.nunique()
```

Nobility

```python
titles = combined['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
titles[np.where(np.isin(titles,['Capt','Col','Major']))]='Millitary'
titles[np.where(np.isin(titles,['Jonkheer','Lady','Master','Sir','Don',"Dona",'the Countess']))]='Nobel'
titles[np.where(np.isin(titles,['Mme','Mrs','Miss']))]='Common Female'
titles[np.where(np.isin(titles,['Mlle','Mr','Ms']))]='Common male'


np.unique(titles,return_counts=True)

combined['Class']=titles
combined.head()
```

Age

```python
bins= np.arange(10,90,10)
combined['age_range']=pd.cut(combined['Age'],bins)
combined.head()
```

ticket count and fare adjustment

```python
tic_dict=dict(combined['Ticket'].value_counts())
combined['Ticket_total']=combined['Ticket'].map(tic_dict)
combined['adjusted_Fare']=combined['Fare']/combined['Ticket_total']


plt.bar(combined['adjusted_Fare'].value_counts().index.values,combined['adjusted_Fare'].value_counts())
plt.show()

combined['fare_range']=pd.cut(combined['adjusted_Fare'],[0,20,40,60,80,150])
combined.head()
```

Family size

```python
combined['family']=combined['Parch']+combined['SibSp']+1
combined.drop(['Ticket','Name','Fare'],axis=1,inplace=True)
combined.head()
```

```python
combined.nunique()
```

# Model Creation for pridiction

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split , RandomizedSearchCV , GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler , LabelEncoder
```

```python
combined.set_index('PassengerId',inplace=True)
combined
```

```python
lebel_enc = combined.copy() 
lebel_enc.drop(['Age','adjusted_Fare'],axis=1,inplace=True)
lebel_enc=lebel_enc.astype(str)
lebel_enc=lebel_enc.apply(LabelEncoder().fit_transform) 
lebel_enc
```

```python
first_te=lebel_enc.copy()
first_te=pd.get_dummies(first_te,columns=['Sex','Embarked','Class']) 
first_te.head()
```

```python
X = first_te.iloc[:len(survived)].values
Y = survived.values


scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.fit_transform(X)


X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=0)


print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
clf =RandomForestClassifier(random_state=0)
clf.fit(X_train,Y_train).score(X_test,Y_test)
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'n_estimators': [50, 100, 200, 300, 500],
    'max_depth': np.arange(3, 500),  
    'min_samples_split': np.arange(1,1000),
    'max_features': ['sqrt', 'log2'],
    'max_samples': np.linspace(0.1, 0.9, 10)
}
randSearch = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=0),
    param_distributions=params,
    scoring='accuracy',
    cv=5,
    random_state=0,
    n_jobs=-1
)

randSearch.fit(X_train,Y_train) 

rand_par = randSearch.best_params_
print("Best Parameters:\n", rand_par)
print("Train Accuracy (CV):", randSearch.best_score_)

# Evaluate on test set
preds = randSearch.predict(X_test)
print("Test Accuracy:", accuracy_score(Y_test, preds))
```

```python
n_estimators = np.linspace(rand_par['n_estimators'] - 10, rand_par['n_estimators'] + 10, 3).astype(int)
min_samples_split = np.arange(rand_par['min_samples_split'] - 2, rand_par['min_samples_split'] + 3).astype(int)
max_samples = np.linspace(rand_par['max_samples'] - 0.01, rand_par['max_samples'] + 0.01, 5)
max_depth = np.arange(rand_par['max_depth'] - 2, rand_par['max_depth'] + 3).astype(int)

# Parameter grid for GridSearchCV
params = {
    'criterion': [rand_par['criterion']],
    'n_estimators': n_estimators,
    'max_depth': max_depth, 
    'min_samples_split': min_samples_split,
    'max_features': [rand_par['max_features']],
    'max_samples': max_samples
}

# Grid search
gridSearch = GridSearchCV(
    estimator=RandomForestClassifier(random_state=0),
    param_grid=params,
    scoring='accuracy',
    cv=5,
    n_jobs=-1
)

# Fit model
gridSearch.fit(X_train, Y_train)

# Results
grid_par = gridSearch.best_params_
print("Best Parameters:\n", grid_par)
print("Train Accuracy (CV):", gridSearch.best_score_)

# Evaluate on test set
preds = gridSearch.predict(X_test)
print("Test Accuracy:", accuracy_score(Y_test, preds))
```
